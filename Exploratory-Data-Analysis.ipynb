## Este notebook es para aplicar el segundo proceso a los datos
### En esta etapa se realiza un analisis de los datos


Primer analisis es buscar por las palabras que mas se repiten

```python
import pandas as pd
data = pd.read_pickle('dtm.pkl')
# hallar la transpuesta de el dataframe no es necesario pero
# es mas sencillo trabajar con las columnas que con las filas
data = data.transpose()
data
```

```python
top_dict = {}
for c in data.columns:
    top = data[c].sort_values(ascending=False).head(30)
    #print(top)
    top_dict[c] = list(zip(top.index,top.values))
top_dict
```

```python
# para imprimir las mas comunes por comediante

for comediantes, top_words in top_dict.items():
    print(comediantes)
    print(','.join([words for words,_count in top_words[0:15]]))
    print('----')
```

## En este punto ya se podría crear la nube de palabras. Pero se puede ver que todavia no queda bien
## Hay palabras muy repetidas en cada uno de los comediantes  como like, im...



```python
# Otra forma de buscar las palabras mas repetidas es usando Counter from collections
from collections import Counter
words = []
for comediantes in data.columns:
    top = [w for (w,count) in top_dict[comediantes]]
    for t in top:
        words.append(t)
words
```

```python
Counter(words).most_common()
```

```python
# hacer que las palabras que se repitan mas de 6 veces sean agregadas como stop words
add_stop_words = [w for w,count in Counter(words).most_common() if count >6]
```

```python
from sklearn.feature_extraction import text
from sklearn.feature_extraction.text import CountVectorizer

data_clean = pd.read_pickle('data_clean.pkl')

stop_words = text.ENGLISH_STOP_WORDS.union(add_stop_words)

cv = CountVectorizer(stop_words=stop_words)
data_cv = cv.fit_transform(data_clean.transcript)
data_stop = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
data_stop.index = data_clean.index
data_stop
```

```python
#para usar despues
import pickle
pickle.dump(cv, open('cv_stop.pkl','wb'))
data_stop.to_pickle('dtm_stop.pkl')
```

```python
# en este punto ya podemos crear nuestra nube de palabras

from wordcloud import WordCloud

wc = WordCloud(stopwords=stop_words, background_color='white',colormap='Dark2',max_font_size=150, random_state=42)
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [16,6]
full_names = ['Ali Wong', 'Anthony Jeselnik', 'Bill Burr', 'Bo Burnham', 'Dave Chappelle', 'Hasan Minhaj',
              'Jim Jefferies', 'Joe Rogan', 'John Mulaney', 'Louis C.K.', 'Mike Birbiglia', 'Ricky Gervais']

for index, comediante in enumerate(data.columns):
    wc.generate(data_clean.transcript[comediante])
    plt.subplot(3,4,index+1)
    plt.imshow(wc, interpolation = 'bilinear')
    plt.axis('off')
    plt.title(full_names[index])

plt.show()
```

### para continuar el analisis es importante saber el vocabulario de cada comediante. para ello, vamos a recorrer el DTM y contar cada palabra que tenga una frecuencia distinta de cero

```python
palabras_unicas = []
for comediante in data.columns:
    unicas = data[comediante].to_numpy().nonzero()[0].size
    print(unicas)
    palabras_unicas.append(unicas)
palabras_unicas
```

```python
# Create a new dataframe that contains this unique word count
data_words = pd.DataFrame(list(zip(full_names, palabras_unicas)), columns=['comedian', 'unique_words'])
data_unique_sort = data_words.sort_values(by='unique_words')
data_unique_sort
```

### Tambien es importante saber el numero de palabras totales utilizadas por cada comediante.




```python
# hallar el numero total de palabras:
total_palabras = []
for comediante in data.columns:
    totalpalabras =  sum(data[comediante])
    total_palabras.append(totalpalabras)

# Create a new dataframe that contains this unique word count
data_words = pd.DataFrame(list(zip(full_names, total_palabras)), columns=['comedian', 'unique_words'])
data_unique_sort = data_words.sort_values(by='unique_words')
data_unique_sort
# Comedy special run times from IMDB, in minutes
# El tiempo de duracion de cada rutina se saca de IMDB
# se puede crear una funcion para saber el tiempo de duracion de cada video y colocarlo en una lista?
# eso se deja de tarea para proximos proyectos

run_times = [60, 59, 80, 60, 67, 73, 77, 63, 62, 58, 76, 79]

# ajustar nuestro dataframe de datos con algunas columnas nuevas para mejor analisis de datos
data_words['total_words'] = total_palabras
data_words['run_times'] = run_times
data_words['words_per_minute'] = data_words['total_words'] / data_words['run_times']
data_words
```
